```python
from pyspark.sql.types import (
    StructType,
    StructField,
    StringType,
    IntegerType,
    DoubleType,
    BooleanType,
    DateType,
    TimestampType
)

df = spark.read.option("header", "true").option("sep", ",").csv(f"/Volumes/demos_standard/public_data/public_incoming/airport_full_20251201.txt")


def spark_type_to_code(dtype):
    if dtype == "string":
        return "StringType()"
    elif dtype == "int":
        return "IntegerType()"
    elif dtype == "bigint":
        return "IntegerType()"
    elif dtype == "double":
        return "DoubleType()"
    elif dtype == "float":
        return "DoubleType()"
    elif dtype == "boolean":
        return "BooleanType()"
    elif dtype == "date":
        return "DateType()"
    elif dtype == "timestamp":
        return "TimestampType()"
    else:
        return "StringType()"

fields = []
for field in df.schema.fields:
    dtype = field.dataType.simpleString()
    fields.append(
        f'    StructField("{field.name}", {spark_type_to_code(dtype)}, {str(field.nullable)})'
    )

print("schema = StructType([\n" + ",\n".join(fields) + "\n])")
```
